### **DATA_SAMPLING**

```python
from datasets import Audio

audio_dataset = common_voice.cast_column("audio", Audio(sampling_rate=16000))

common_voice_train = audio_dataset


#common_voice_val = audio_dataset["validation"]
#common_voice_test = audio_dataset["test"]
```

```python
from transformers import WhisperFeatureExtractor, WhisperTokenizer, WhisperProcessor

# Load the feature extractor for the pre-trained model
feature_extractor = WhisperFeatureExtractor.from_pretrained(model_name_or_path)
# Load the tokenizer for the pre-trained model and specified language/task
tokenizer = WhisperTokenizer.from_pretrained(model_name_or_path, language=language, task=task)
# Load the processor which combines the feature extractor and tokenizer
processor = WhisperProcessor.from_pretrained(model_name_or_path, language=language, task=task)
```

### **DATA SPLIT FOR READINESS TO PUSH BACK TO HUGGINGFACE**

```python
# Set a seed for reproducibility of random operations
SEED = 42  # choose any fixed integer

# We split the dataset into multiple batches for processing and potential parallelisation.
# Initially, we split the dataset into 4 batches.
train_batch_1 = common_voice_train
# train_batch_2 = common_voice_val # commented out as validation data is not used
# test_batch_A = common_voice_test # commented out as test data is not used

# Split the first training batch into two halves
train_batch_A_split = train_batch_1.train_test_split(test_size=0.5, seed=SEED)

# Reassign the split halves to train_batch_1 and train_batch_2
train_batch_1 = train_batch_A_split["train"]
train_batch_2 = train_batch_A_split["test"]

# Split the two batches further into 4 more batches
train_batch_a_split = train_batch_1.train_test_split(test_size=0.5, seed=SEED)
train_batch_b_split = train_batch_2.train_test_split(test_size=0.5, seed=SEED)

# Reassign the split halves to train_batch_one, train_batch_two, train_batch_three, and train_batch_four
train_batch_one = train_batch_a_split["train"]
train_batch_two = train_batch_a_split["test"]
train_batch_three = train_batch_b_split["train"]
train_batch_four = train_batch_b_split["test"]


# Split the four batches further into 8 batches
train_batch_aa_split = train_batch_one.train_test_split(test_size=0.5, seed=SEED)
train_batch_bb_split = train_batch_two.train_test_split(test_size=0.5, seed=SEED)
train_batch_cc_split = train_batch_three.train_test_split(test_size=0.5, seed=SEED)
train_batch_dd_split = train_batch_four.train_test_split(test_size=0.5, seed=SEED)

# Reassign the split halves to the final 8 batches
train_batch_un = train_batch_aa_split["train"]
train_batch_do = train_batch_aa_split["test"]
train_batch_thro = train_batch_bb_split["train"]
train_batch_quat = train_batch_bb_split["test"]
train_batch_cinq = train_batch_cc_split["train"]
train_batch_sss = train_batch_cc_split["test"]
train_batch_sept = train_batch_dd_split["train"]
train_batch_ng = train_batch_dd_split["test"]
```

```python
def prepare_dataset(batch):
    """
    Prepares a single batch of the dataset for training by processing audio and text data.

    Args:
        batch (dict): A dictionary representing a single data sample with 'audio' and 'sentence' keys.

    Returns:
        dict: The processed batch with added 'input_length', 'input_features', 'labels', and 'labels_length' keys.
    """
    # load audio data
    audio = batch["audio"]

    # compute input length
    batch["input_length"] = len(batch["audio"])

    # compute log-Mel input features from input audio array
    batch["input_features"] = feature_extractor(audio["array"], sampling_rate=audio["sampling_rate"]).input_features[0]

    # encode target text
    batch["labels"] = tokenizer(batch["sentence"]).input_ids

    # labels length
    batch["labels_length"] = len(batch["labels"])
    return batch
```

### **PROCESS AND COMMIT TO HUGGINGFACE**

```python
# Create a list of all your training batches
training_batches = [
    train_batch_un,
    train_batch_do,
    train_batch_thro,
    train_batch_quat,
    train_batch_cinq,
    train_batch_sss,
    train_batch_sept,
    train_batch_ng
]

# Create a list of the corresponding suffixes for the hub names
suffixes = ["11", "22", "33","44","55","66","77","88"]

# Loop through both lists simultaneously using zip()
for batch, suffix in zip(training_batches, suffixes):
    preprocessed_batch = batch.map(prepare_dataset)
    preprocessed_batch.push_to_hub(f"Username/swahili_medium_validation{language}data_{suffix}")
```
